```python
from bs4 import BeautifulSoup
import requests
```

```python
year = '2016'
make = 'cadillac'
model = 'ats'
submodel = ''
parts = 'bumpers'

url = "https://www.carid.com/"+year+"-"+make+"-"+model+"-"+parts+"/"
print(url)
```

```python
# response = requests.get('https://www.napaautocare.com/estimator.aspx')
# soup = BeautifulSoup(response.text,"lxml")

response = requests.get(url)
soup = BeautifulSoup(response.text,"lxml")
```

```python
print(soup.prettify)
```

```python
soup.title()
```

```python
all_spans = soup.find_all("span")
all_spans
```

```python
all_prices = soup.find_all('span',{'class':'lst_prc'} )
all_prices
```

```python
import numpy as np

cost = []
for price in all_prices:
    #cost.append (price.get('$'))
    cost+=(str(price.string).replace('$','').replace(' ','').replace(',','').split('-'))
        
print(cost)

cost_int = [float(c) for c in cost]

mean_cost = np.mean(cost_int)

big_cost = [c for c in cost_int if c >= mean_cost]

mean_big_cost = np.mean(big_cost)

mean_big_cost
```
